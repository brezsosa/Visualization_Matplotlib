```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data)
df['Class'] = iris.target
df.columns = ['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid', 'class']
df.head()
```

```python
ndf = df.drop('class', axis = 'columns')
ndf.head()
```

```python
plt.figure()
plt.scatter(ndf[ndf.columns[0]], ndf[ndf.columns[1]], marker = 'o', alpha = 0.5)
plt.xlabel(ndf.columns[0])
plt.ylabel(ndf.columns[1])
```

```python
import seaborn as sns

sns.pairplot(ndf)
```

```python
def scatter(colnames):
    plt.figure(figsize = [10,10])
    plt.scatter(df[colnames[0]], df[colnames[1]], alpha = 0.5)
    plt.xlabel(colnames[0])
    plt.ylabel(colnames[1])
    plt.title('{} by {}'.format(colnames[1], colnames[0]))
    plt.show()
```

```python
scatter(['sepal_len', 'sepal_wid'])
```

```python
@interact(colnames = widgets.SelectMultiple(options = ndf.columns, value = (ndf.columns[0],ndf.columns[1]), description = 'Features'))
def scatter(colnames):
    plt.figure(figsize = [5,5])
    plt.scatter(ndf[colnames[0]], ndf[colnames[1]], alpha = 0.5)
    plt.xlabel(colnames[0])
    plt.ylabel(colnames[1])
    plt.title('{} by {}'.format(colnames[1], colnames[0]))
    plt.show();
```
