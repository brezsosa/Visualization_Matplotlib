# Matplotlib

```python
# We import the scritping layer of matplotlib

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Scatterplot

```python
## Lets use an array of random numbers to create a scatterplot usign plt.scatter()

x = np.random.rand(5)
y = x

plt.figure()  # First we create a figure to plot in
plt.scatter(x, y) # We create the scatter plot 
plt.show()
```

```python
# We can customize the plot if we use other functions on plt and if we add parameters

x = np.random.rand(5)
y = x

plt.figure()
plt.scatter(x[:2], y[:2], c = ['red', 'red'], s = 45, label = 'Group1')
plt.scatter(x[2:], y[2:], c =['green', 'green', 'green'], s = 80, label = 'Group2')
plt.xlabel('Some independent variable')
plt.ylabel('Some dependent variable')
plt.title('Some title')
plt.legend()  ## Uses the parameter 'labels' in the scatter() function
plt.show()
```

## Lineplots

```python
# Line plots are constructed using the plot() function. It uses the index of the series by default as the x-axis.

line = np.array([1,2,3,4,5,6,7,8,])
quadratic = line ** 2

plt.figure()
plt.plot(line, '-o', quadratic, '-o') # We can pass as many Series as we want. '-o' means line and dots
plt.plot(np.array([22, 44, 55]), '--r') # '--r' means red dashed line
plt.show()
```

```python
## Adding title and legends

line = np.array([1,2,3,4,5,6,7,8,])
quadratic = line ** 2

plt.figure()
plt.legend(['Competition', 'Baseline', 'Us'])
plt.title('A title')
plt.xlabel('Some ind data')
plt.ylabel('Some dep data')
plt.plot(line, '-o', quadratic, '-o') # We can pass as many Series as we want. '-o' means line and dots
plt.plot(np.array([22, 44, 55]), '--r') # '--r' means red dashed line
plt.show()
```

```python
## Say we want to fill in spaces between lines


plt.figure()
plt.legend(['Competition', 'Baseline', 'Us'])
plt.title('A title')
plt.xlabel('Some ind data')
#brezso590
#2528
plt.ylabel('Some dep data')
plt.plot(line, '-o', quadratic, '-o') # We can pass as many Series as we want. '-o' means line and dots
plt.plot(np.array([22, 44, 55]), '--r') # '--r' means red dashed line
plt.gca().fill_between(range(len(line)), line, quadratic, facecolor = 'blue', alpha = 0.25) ## 'range' gives the x axis, 'line' and 'quadratic' give the lower and upper bound (respectively)
plt.show()
```

```python
## Working with dates. Lets include a time series as the x axis of our lineplot (the previous default was the index of the series)
timeseries = np.arange('2023-06-01', '2023-06-09', dtype = 'datetime64[D]')  ## Time series len should be the same as the lenght of the lines plotted to work properly as an x axis

plt.figure()
plt.plot(timeseries, line, '-o', timeseries, quadratic, '-o')
plt.show()
```

```python
## To change the position of the ticks in the x axis:


plt.figure()
plt.plot(timeseries, line, '-o', timeseries, quadratic, '-o')
x = plt.gca().xaxis # We create a variable to acces the x axis elements

for i in x.get_ticklabels():   ## We iterate over every ticklabel and change the rotation angle
    i.set_rotation(45)

plt.show()
```

```python
# The figure size is pass a tuple parameter inside the figure() function

plt.figure(figsize=(10,8))
plt.plot(timeseries, line, '-o', timeseries, quadratic, '-o')
x = plt.gca().xaxis # We create a variable to acces the x axis elements

for i in x.get_ticklabels():   ## We iterate over every ticklabel and change the rotation angle
    i.set_rotation(45)

plt.show()
```

## Bar Charts

```python
## Lets create a bar chart with linear data

data = np.array([1,2,3,4,5,6,7,8])
xvals = range(len(data))

plt.figure()
plt.bar(xvals, data, width= 0.3) ## Bar function needs a value for the x-axis
plt.show()
```

```python
## To add another bar we need to 'correct' the xvals to insert it
exponential = data ** 2
xvals2 = []
for i in xvals:
    xvals2.append(i + 0.3)

plt.bar(xvals, data,width = 0.3)
plt.bar(xvals2, exponential, width = 0.3, color = 'red')
plt.show()
```

```python
## Or we can stack the bars

xvals2 = xvals
plt.bar(xvals, data,width = 0.3)
plt.bar(xvals2, exponential, width = 0.3, color = 'red', bottom = data)
plt.show()
```

```python
## We can create horizontal bar charts with barh()
plt.barh(xvals, data,height = 0.3)
plt.barh(xvals2, exponential, height = 0.3, color = 'red', left = data)  ## Note that widht now is height and bottom now is left
plt.show()
```

```python
# If we want to drop the frame 

plt.title('Stacked bars \nas example barchart')
plt.bar(xvals, data,width = 0.3)
plt.bar(xvals2, exponential, width = 0.3, color = 'red', bottom = data)

x = plt.gca()

for i in x.spines.values():
    i.set_visible(False)

plt.show()
```

```python
## Change the color of one bar
data = np.array([1,2,3,4,5,6,7,8])
xvals = np.arange(len(data))
names = ['Alba', 'Rosa', 'Maga', 'Eva', 'Flor', 'Clara', 'Cami', 'Lau']

plt.figure()
bars = plt.bar(xvals, data, width= 0.8, align = 'center', linewidth = 0, color = 'lightslategray') ## Bar function needs a value for the x-axis
bars[0].set_color('darkred')
plt.xticks(xvals, names)
x = plt.gca()
for i in x.spines.values():
    i.set_visible(False)
plt.show()
```

```python
## To get rid of the yticks and insert the y value in each bar

plt.figure()
bars = plt.bar(xvals, data, width= 0.8, align = 'center', linewidth = 0, color = 'lightslategray') ## Bar function needs a value for the x-axis
bars[0].set_color('#1F77B4')
plt.xticks(xvals, names)
plt.yticks([])
x = plt.gca()
for i in x.spines.values():
    i.set_visible(False)
plt.title('Some title \nwith two lines')

for bar in bars:
    height = bar.get_height()
    plt.gca().text(bar.get_x() + bar.get_width() / 2, height - 0.5, str(int(height)) + '%', ha = 'center', color = 'w', fontsize = 9)
plt.show()
```

## Subplots

```python
# We can create subplots using the subplot() function. First arg: rows, Second arg: cols, Third argument: position
data = np.array([1,2,3,4,5,6])
plt.figure()
ax1 = plt.subplot(121)
plt.plot(data, '-o')
plt.subplot(122, sharey = ax1) ## We use sharey to share the same y axis between the charts
plt.plot(data**2,'-o')
plt.show()
```

```python
# We can get rid of the chartjunk

plt.figure()
ax = plt.subplot(121)
plt.plot(data, '--')
for i in plt.gca().spines.values():
    i.set_visible(False)
plt.subplot(122, sharey = ax)
plt.plot(data **2, '--')
for i in plt.gca().spines.values():
    i.set_visible(False)
plt.show()

```

```python
## We can create a grid os multple scatterplots (SPLOM - scatterplotmatrix), to compare several columns of a df

import pandas as pd
df = pd.read_csv('./data/mpg.csv')
df = df.drop('Unnamed: 0', axis= 'columns')
df.columns = [i.upper() for i in df.columns]
df.head()
```

```python
# We change the datatpe to floats to further construct the scatterplots

df['CTY'] = df['CTY'].astype(float)
df['DISPL'] = df['DISPL'].astype(float)
df['HWY'] = df['HWY'].astype(float)
```

```python
# Creation of the grid

cols = ['DISPL', 'CTY', 'HWY']
plt.figure()
fig, axs = plt.subplots(len(cols), len(cols), figsize = (10,10))

# We need to iterate over every row and every column of the SPLOM to construct a scatterplot at each position:
for i in range(len(cols)):
    for j in range(len(cols)):
        
        #This constructs the SPLOM:
        axs[i,j].scatter(df[cols[i]], df[cols[j]], s = 10)
        
        ## This turns off the axis:
        axs[i,j].get_xaxis().set_visible(False)
        axs[i,j].get_yaxis().set_visible(False)

        ## These if statements set visible and label only the edges of the SPLOM:
        if i == len(cols) - 1:
            axs[i,j].get_xaxis().set_visible(True)
            axs[i,j].set_xlabel(cols[j])
        if j == 0:
            axs[i,j].get_yaxis().set_visible(True)
            axs[i,j].set_ylabel(cols[i])
    
```

```python
plt.scatter(df[cols[0]], df[cols[1]])
plt.xlabel(cols[0])
plt.ylabel(cols[1])
```

```python
import seaborn as sns
ndf = df[cols]
sns.pairplot(ndf);
```

We can conclude as a first exploratory approximation that there seems to be a linear relationship between 'HWY' and 'CTY' variables.

```python
## We can alternatively construct a grid with much more flexibility using this approach:

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, figsize = (5,5), sharex = False, sharey = False)
ax1.plot(data, '--')
ax2.plot(exponential, '-')
ax3.bar(range(5), np.array([0,1,2,3,4]))
ax4.bar(range(7), np.array([0,15,25,32,54,45,38]));
```

## Seaborn

```python
import seaborn as sns   ## Seaborn is a plotting library built on top of matplotlib. Much of the functionality of seaborn uses matplotlib artists to create plots
```

```python
#Lets create two series of random data

np.random.seed(123)

v1 = pd.Series(np.random.normal(0,10,1000), name = 'v1')  ## The arguments in np.random.normal are: (mean, std, observations)
v2 = pd.Series(v1 * 2 + np.random.normal(60,15,1000), name = 'v2')
```

```python
## Creates a distribution plot

v3 = pd.concat([v1, v2])
plt.figure()
sns.displot(v3, kde_kws= {'color': 'Navy'});
```

```python
## Joint plot

sns.jointplot((v1, v2), alpha = 0.4);
```
